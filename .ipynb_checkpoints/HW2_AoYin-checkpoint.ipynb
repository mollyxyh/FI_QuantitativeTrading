```python
from numpy import *
from pandas import *
import statsmodels.tsa.stattools as ts
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

##import the data and extract the specific columns
df = pandas.read_excel('CMT-all.xlsx')
df.set_index('Date', inplace=True)
df = df.loc[:,['3M', '2Y','5Y','7Y','10Y','30Y']]
print df.head()
```

```python
##construct the different subsets
df.index = to_datetime(df.index)
sample_1 = df[(df.index >= to_datetime('20130101')) & (df.index <= to_datetime('20141231'))]#2013-2014 history
sample_2 = df[(df.index >= to_datetime('20150101')) & (df.index <= to_datetime('20161231'))]#2015-2016 history
print ('sample1: \n') 
print sample_1.head()
print '\n'
print ('sample2: \n') 
print sample_2.head()
```

1.	Perform PCA decomposition using Sample1, 5 pts

```python
%matplotlib inline
C = sample_1.cov()
# compute the eigenvalue and eigenvector, orderd by descending
evals,evecs = linalg.eig(C)
indices = argsort(evals)
indices = indices[::-1]
evecs = evecs[:,indices]
evals = evals[indices]
for i in range(shape(evecs)[1]):
    evecs[:,i] / linalg.norm(evecs[:,i]) * sqrt(evals[i])
print ('eigenvalueï¼š\n')
print evals
print '\n'
print ('PCA: 3M    2Y    5Y    7Y   10Y   30Y\n')
print evecs

# produce new data matrix
x = dot(sample_1, evecs)
plt.plot(x[:,0:3])
plt.title('The first 3 principal')
```

2.	Compute Half-Life or Augmented Dickey-Fuller (ADF) test statistic for:
  1.	FLY in Sample1, 5 pts
  2.	PCA FLY in Sample1, 5 pts 
  3.	& 4: repeat 1 & 2 in Sample2 (5 pts each, note that weights of the PCA FLY were determined in Sample1)


```python
##FLY in sample1
FLY_1 = sample_1['2Y']-sample_1['5Y']+sample_1['10Y']
adf_fly_1 = ts.adfuller(FLY_1, 1)#Augmented Dickey-Fuller (ADF) test

##FLY in sample 2
FLY_2 = sample_2['2Y']-sample_2['5Y']+sample_2['10Y']
adf_fly_2 = ts.adfuller(FLY_2, 1)#Augmented Dickey-Fuller (ADF) test 

print ('ADF for FLY in Sample 1ï¼š\n')
print adf_fly_1
print '\n'
print ('ADF for FLY in Sample 2: \n')
print adf_fly_2
```

  From the results of the Augmented Dickey-Fuller (ADF) tests, we can that for sample 1 and  the sample 2, the FLYs do not pass the test, which means there is nonstationarity for each series.

```python
##PCA FLY 
a = np.array([[evecs[1][0], evecs[4][0]], [evecs[1][1], evecs[4][1]]])
b = np.array([evecs[2][0], evecs[2][1]])
w = linalg.solve(a, b)#compute the weights
PCA_FLY_1 = w[0]*sample_1['2Y']-sample_1['5Y']+w[1]*sample_1['10Y']
adf_PCAfly_1 = ts.adfuller(PCA_FLY_1, 1)#Augmented Dickey-Fuller (ADF) test
PCA_FLY_2 = w[0]*sample_2['2Y']-sample_2['5Y']+w[1]*sample_2['10Y']
adf_PCAfly_2 = ts.adfuller(PCA_FLY_2, 1)#Augmented Dickey-Fuller (ADF) test

print ('ADF for PCA FLY in Sample 1ï¼š\n')
print adf_PCAfly_1
print '\n'
print ('ADF for PCA FLY in Sample 2: \n')
print adf_PCAfly_2
```

  From the results of the Augmented Dickey-Fuller (ADF) tests, we can that for sample 1, the p-value of the PCA FLY is very small. That's to say the result is significant and this series is stationary. But for the sample 2, the PCA FLY does not pass the test, which means there is nonstationarity.

3.	Perform Cointegration analysis either by Box-Tiao or Chou-Ng in Sample1, 20 pts.  Alternatively, compute a COINT FLY from a level regression, 5 pts

```python
###We use two ways to compute the COINT FLY

##First we perform cointegration analysis
sample1_post = sample_1[0:(len(sample_1)-1)]
sample1_next = sample_1[1:len(sample_1)]
A = linalg.lstsq(sample1_post,sample1_next)[0]#product the matrix A that calibrates the relationship between Yt and Yt-1
print ('relationship between Yt and Yt-1: \n')
print A
print '\n'
Q = dot(dot(dot(linalg.inv(C),A),C),A.T)#Q is a measure of relative predictability

##Perform Cointegration analysis
CCA_evals,CCA_evecs = linalg.eig(Q)
CCA_indices = argsort(CCA_evals)
CCA_indices = CCA_indices[::-1]
CCA_evecs = CCA_evecs[:,CCA_indices]
CCA_evals = CCA_evals[CCA_indices]
for i in range(shape(CCA_evecs)[1]):
    CCA_evecs[:,i] / linalg.norm(CCA_evecs[:,i]) * sqrt(CCA_evals[i])
print ('eigenvalueï¼š\n')
print CCA_evals
print '\n'
print ('factor loadings of the Cointegration Model\n')
print CCA_evecs

# produce new data matrix
fac = dot(sample_1,CCA_evecs)
plt.plot(fac[:,0:3])
plt.title('The first 3 factors')
```

4.	Compute Half-Life or Augmented Dickey-Fuller (ADF) test statistic for:
  1.	COINT FLY in Sample1, 5pts
  2.	COINT FLY in Sample2 (note, that weights of the COINT FLY were chosen using Sample1), 5pts


```python
##COINT FLY 
#a_ = np.array([[CCA_evecs[1][4], CCA_evecs[4][4]], [CCA_evecs[1][5], CCA_evecs[4][5]]])
#b_ = np.array([CCA_evecs[2][4], CCA_evecs[2][5]])
#w_coint = linalg.solve(a_,b_)
w_coint = [CCA_evecs[1][5]/CCA_evecs[2][5], CCA_evecs[4][5]/CCA_evecs[2][5]]#compute the weights

coint_FLY_1 = w_coint[0]*sample_1['2Y']-sample_1['5Y']+w_coint[1]*sample_1['10Y']
adf_cointfly_1 = ts.adfuller(coint_FLY_1, 1)#Augmented Dickey-Fuller (ADF) test
coint_FLY_2 = w_coint[0]*sample_2['2Y']-sample_2['5Y']+w_coint[1]*sample_2['10Y']
adf_cointfly_2 = ts.adfuller(coint_FLY_2, 1)#Augmented Dickey-Fuller (ADF) test
print w_coint
print ('ADF for COINT FLY in Sample 1 By CCAï¼š\n')
print adf_PCAfly_1
print '\n'
print ('ADF for COINT FLY in Sample 2 By CCA: \n')
print adf_PCAfly_2
```

```python
##second we do level regression 
lr = LinearRegression()
lr.fit(sample_1[['2Y','10Y']], sample_1['5Y'])
w_ = lr.coef_ #get the weights
COINT_FLY_1 = w_[0]*sample_1['2Y']-sample_1['5Y']+w_[1]*sample_1['10Y']
adf_COINTfly_1 = ts.adfuller(COINT_FLY_1, 1)#Augmented Dickey-Fuller (ADF) test
COINT_FLY_2 = w_[0]*sample_2['2Y']-sample_2['5Y']+w_[1]*sample_2['10Y']
adf_COINTfly_2 = ts.adfuller(COINT_FLY_2, 1)#Augmented Dickey-Fuller (ADF) test

print ('ADF for COINT FLY in Sample 1 By regressionï¼š\n')
print adf_COINTfly_1
print '\n'
print ('ADF for COINT FLY in Sample 2 By regression: \n')
print adf_COINTfly_2
```

  Similar to the results we get in the PCA part, the COINT FLYs we derive from both two ways also show that for sample 1 the series is stationary but not for sample2.
