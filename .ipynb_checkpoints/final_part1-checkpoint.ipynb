```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read files into pandas dateframe
df = pd.read_csv('closes_Quandl_Rolling_EDs.csv')
df.set_index('Date', inplace=True)
df.index = pd.to_datetime(df.index)
df = df.rename(columns=lambda x: x.split('_')[1], inplace=False)

#vol = pd.read_csv('Final_CM_ED_Vol.csv')
```

```python
def mu_t(mu, t):
    return mu

def integrated_mu(mu, t):
    result = integrate.quad(lambda x: mu_t(mu, x), 0,t)
    return result[0]

def mean_r(mu, var, t, dist = 'Q'): 
    m = r0+mu*t
    if (dist == 'F'):
        m -= var * pow(t,2)
    return m

#int{r(u)du}:
def mean_int_r(mu, t):
    return r0*t+mu*pow(t,2)/2.

def var_int_r(var, t):
    return var*pow(t,3)/3

def A(mu, var, tau):
    return var*pow(tau,3)/6. - mu*pow(tau,2)/2.

def B(tau):
    return tau

def zero(mu, var, t0, T, r_t0):
    tau = T-t0
    return np.exp(A(mu, var, tau)-B(tau)*r_t0)

def fwd_rate(mu, var, t_fix, T0, T1, dist = 'Q', acrual=0.25):
    A_diff = A(mu, var, T0-t_fix) - A(mu, var, T1-t_fix)
    B_diff = B(T0-t_fix) - B(T1-t_fix)
    m = A_diff - B_diff * mean_r(mu, var, t_fix, dist)
    v = var * t_fix * pow(B_diff, 2)
    e = np.exp(m+v/2.)
    return 1/acrual * (e-1)
```

```python
r0 = 0.0135
mu = 0.005
vol = 0.0075
var = pow(vol,2)

x = range(1,21)
term = [xx/4. for xx in x]

rate_Q = [fwd_rate(mu, var, t, t, t+0.25, 'Q')*100 for t in term]

def plot2(dt,frame,x,y):
    f = plt.figure()
    ax = f.add_subplot(111)
    ax.yaxis.tick_right()  
    crv = frame.loc[dt]
    plots = [plt.plot(x,crv.values,label=dt,linestyle="dotted",marker='o'),plt.plot(x,y,label='Model')]
    plt.legend(loc='lower right')    
    plt.show()
    
dt = '2017-04-04'
plot2(dt,df,term,rate_Q)
```

```python
params = [mu, var]
def residuals_TSM(params):
    rate_Q = [fwd_rate(params[0], params[1], t,t,t+0.25,'Q')*100 for t in term]
    return sum((df.values[0]-rate_Q)**2)

from scipy.optimize import least_squares
#opt = leastsq(residuals_TSM, params)
opt = least_squares(residuals_TSM, params, bounds=((0,0),(np.inf, 1))).x

rate_opt = [fwd_rate(opt[0], opt[1], t,t,t+0.25,'Q')*100 for t in term]
plot2(dt,df,term,rate_opt)
```

```python
import statsmodels .api as sm

def alpha(t,l):
    return ((1-np.exp(-l*t))/(l*t))

lam = 0.0609
term = np.linspace(3,3*20,20)
b2 = [alpha(t,lam) for t in term]
b3 = [alpha(t,lam)-np.exp(-lam*t) for t in term]

#create the array to store betas
beta_1 = []
beta_2 = []
beta_3 = []

#define the independent variables
x = np.array([b2+b3]).reshape(2,-1).T
x = sm.add_constant(x)

#fit regression model --- example for 2017-04-04
dt = '2017-04-04'
yt = df.loc[dt]
results = sm.OLS(yt,x).fit()
plots = [plt.plot(term, yt, label = 'CMT' + dt), plt.plot(term, results.fittedvalues, label = 'NS' + dt)]
plt.legend(loc = 'lower right')
plt.show()
print (results.summary())

for t in df.index:
    y = df.loc[t]
    result = sm.OLS(y,x).fit()
    beta_1.append(result.params[0])
    beta_2.append(result.params[1])
    beta_3.append(result.params[2])

mean_beta1 = np.mean(np.nan_to_num(np.array(beta_1)))
mean_beta2 = np.mean(np.nan_to_num(np.array(beta_2)))
mean_beta3 = np.mean(np.nan_to_num(np.array(beta_3)))
std_beta1 = np.std(np.nan_to_num(np.array(beta_1)))
std_beta2 = np.std(np.nan_to_num(np.array(beta_2)))
std_beta3 = np.std(np.nan_to_num(np.array(beta_3)))
```

```python
# construct PCAs from the levels of yields.
cov = df.cov()
eig_val_cov, eig_vec_cov = np.linalg.eig(cov) # calculate eigenvalues and eigenvectors of cov_1
for ev in eig_vec_cov: # assert the correctness of calculation
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev)) # make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_val_cov[i]), eig_vec_cov[:,i]) for i in range(len(eig_val_cov))]
eig_pairs.sort(key=lambda x: x[0], reverse=True) # sort the (eigenvalue, eigenvector) tuples from high to low

matrix_w = np.hstack((eig_pairs[0][1].reshape(20,1), eig_pairs[1][1].reshape(20,1), eig_pairs[2][1].reshape(20,1)))
print 'First three components of key rates: \n', matrix_w, '\n'

# plot time series of the first 3 principal components
ts_3c = np.dot(df.values, matrix_w)
plt.plot(ts_3c)
plt.xlabel('time')
plt.ylabel('time series of the first three components')
plt.title('The first 3 principal components of the yield curve')
plt.show()
```

```python
def mgf(mu, var, t_fix,T0,T1,k,dist = 'Q'):
    A_diff = A(mu, var, T0-t_fix) - A(mu, var, T1-t_fix)
    B_diff = B(T0-t_fix) - B(T1-t_fix)
    m = A_diff - B_diff * mean_r(mu, var, t_fix,dist)
    v = var * t_fix * pow(B_diff,2)
    return np.exp(k*(2*m+k*v)/2.)

def fwd_rate_mean(mu, var, t_fix, T0, T1, dist='Q', acrual=0.25):
    m1 = mgf(mu, var, t_fix, T0, T1, 1, dist)
    return 1/acrual*(m1-1)

def fwd_rate_vol(mu, var, t_fix, T0, T1, dist='Q', acrual=0.25):
    m2 = mgf(mu, var, t_fix, T0, T1, 2, dist)
    m1 = mgf(mu, var, t_fix, T0, T1, 1, dist)
    e2 = m2 - pow(m1, 2)
    return 1/acrual*pow(e2/t_fix,0.5)
```

```python
vol = [fwd_rate_vol(mu, var, t,t,t+0.25,'F')*10000/16 for t in range(1,20)]
plt.plot(vol, label = 'fwd libor volatility in bps')
plt.legend(loc = 'best')
plt.show()
```
