```python
import pandas as pd
import numpy as np
from pandas.stats.api import ols
import matplotlib.pyplot as plt
from scipy.optimize import minimize
#from sklearn.mixture import GaussianMixture
```

```python
# import output data from homework3
data = pd.read_csv('DataSet3.csv')
data.set_index('0', inplace=True)
data = data.dropna() # remove na values

# split the data into training sample, cross-validation sample and test sample
sampleA = data.loc['6/1/11':'1/31/14',:]
sampleB = data.loc['6/2/14':'5/29/15',:]
sampleC = data.loc['6/1/15':'5/31/16',:]

# print first five rows of three samples
print 'sampleA: \n', sampleA.head(), '\n'
print 'sampleB: \n', sampleB.head(), '\n'
print 'sampleC: \n', sampleC.head(), '\n'
```

### Training Sample Processing

```python
# extract vector element from sampleA
y2 = sampleA.iloc[:,7]
y3 = sampleA.iloc[:,11]
y4 = sampleA.iloc[:,15]
y475 = sampleA.iloc[:,18]

# build cointegrated vectors using OLS (signal1)
c1 = ols(y=y3, x=y2)
c2 = ols(y=y4, x=y3)
c3 = ols(y=y475, x=y4)

cv1 = c1.resid
cv2 = c2.resid
cv3 = c3.resid

# build AR(1) on signal1 cv1-3
ar1 = ols(y=cv1, x=cv1.shift(1))
ar2 = ols(y=cv2, x=cv2.shift(1))
ar3 = ols(y=cv3, x=cv3.shift(1))
print 'Summary of AR(1) for cv1', ar1, '\n'
print 'Summary of AR(1) for cv2', ar2, '\n'
print 'Summary of AR(1) for cv3', ar3, '\n'

# calculate half-lives for signal1 cv1-3
hl1 = np.log(0.5)/(np.log(ar1.beta[0]))
hl2 = np.log(0.5)/(np.log(ar2.beta[0]))
hl3 = np.log(0.5)/(np.log(ar1.beta[0]))
print 'half-lives for cv1-3 under signal1: ', hl1, hl2, hl3
```

For each cointegrated vector in signal2, the alphas chosen to make HL close to 10 are 1/28, 2/33 and 1/16 respectively. And the corresponding spans are 55, 32 and 31 respectively. 

```python
# build signal2 (span = 2/a -1) (a=2/(1+span))
s1 = cv1 - cv1.ewm(span=55).mean() # a=1/28
s2 = cv2 - cv2.ewm(span=32).mean() # a=2/33
s3 = cv3 - cv3.ewm(span=31).mean() # a=1/16

# build AR(1) on signal2 s1-3
ar1_ = ols(y=s1, x=s1.shift(1))
ar2_ = ols(y=s2, x=s2.shift(1))
ar3_ = ols(y=s3, x=s3.shift(1))
print 'Summary of AR(1) for s1', ar1_, '\n'
print 'Summary of AR(1) for s2', ar2_, '\n'
print 'Summary of AR(1) for s3', ar3_, '\n'

# calculate half-lives for signal2 s1-3
hl1_ = np.log(0.5)/(np.log(ar1_.beta[0]))
hl2_ = np.log(0.5)/(np.log(ar2_.beta[0]))
hl3_ = np.log(0.5)/(np.log(ar3_.beta[0]))
print 'half-lives for s1-3 under signal2: ', hl1_, hl2_, hl3_
```

### Cross-validation Sample Processing
*We will use Cross-Validation Sample B to define a trading strategy and solve for the optimal weight of the Mixture.*

```python
# extract vector element from sampleB
Y2 = sampleB.iloc[:,7]
Y3 = sampleB.iloc[:,11]
Y4= sampleB.iloc[:,15]
Y475 = sampleB.iloc[:,18]

# calculate signals1&2 in sampleB using parameters obtained from sampleA
CV1 = Y3-Y2*c1.beta[0]-c1.beta[1]
CV2 = Y4-Y3*c2.beta[0]-c2.beta[1]
CV3 = Y475-Y4*c3.beta[0]-c3.beta[1]
S1 = CV1 - CV1.ewm(span=55).mean()
S2 = CV2 - CV2.ewm(span=32).mean()
S3 = CV3 - CV3.ewm(span=31).mean()

# concate two signals to feed in GMM algo
sample1 = pd.concat([CV1, S1], axis=1)
sample2 = pd.concat([CV2, S2], axis=1)
sample3 = pd.concat([CV3, S3], axis=1)

# train GMM models
#g = GaussianMixture(n_components=2)
#g.fit(sample1) 
#w1 = g.weights_
#g.fit(sample3)
#w3 = g.weights_
```

```python
# show a specific example of the first cointegrated vector on sampleB when mixture weight=0.5
weight = 0.5
mix = CV1*weight + S1*(1-weight)
trading = [0,0] # trading signals
for i in range(2, len(sampleB)-5):
    c = mix[i]
    ar = ols(y=mix[:i], x=mix[:i].shift(1)) # build dynamic AR(1) model
    for _ in range(5): # predict rates in five days
        c = c*ar.beta[0]+ar.beta[1]
    if c-mix[i]>0.00001: # If |Expected Gains(t)|>0.1 basis point, your trade is initiated
        trading.append(1)
    else:
        trading.append(0)

realize = [] # store daily gain in a list
for i in range(len(trading)):
    realize.append(trading[i]*(mix[i+5]-mix[i]))
realize_cumsum = np.cumsum(realize) # calculate the cumsum of daily gains

plt.plot(realize_cumsum) # plot the gains of specific example
plt.title('Realized gain with the weight of Gaussian Mixture equal to 0.5')
plt.xlabel('days')
plt.ylabel('Gains')
plt.show()
```

Then I define a general function to calculate realized gains on cross-validation sample for the first cointegrated vector, apply it to a range of weights and select an optimal weight that maximizes realized gains on cross-validation return. The reason to do the manual search is because till now I haven't found an appropriate optimization packages for this problem in python.

```python
def realized_gains(weight, CV, S):
    mix = CV*weight + S*(1-weight)
    trading = [0,0]
    for i in range(2, len(sampleB)-5):
        c = mix[i]
        ar = ols(y=mix[:i], x=mix[:i].shift(1))
        for _ in range(5):
            c = c*ar.beta[0]+ar.beta[1]
        if c-mix[i]>0.00001:
            trading.append(1)
        else:
            trading.append(0)

    realize = 0
    for i in range(len(trading)):
        realize += trading[i]*(mix[i+5]-mix[i])
    return realize

options = np.linspace(0,1,50) # define a range of weights
output1 = [realized_gains(item, CV1, S1) for item in options]
output2 = [realized_gains(item, CV2, S2) for item in options]
output3 = [realized_gains(item, CV3, S3) for item in options]
```

```python
pos1 = output1.index(max(output1))
pos2 = output2.index(max(output2))
pos3 = output3.index(max(output3))
optw1 = options[pos1]
optw2 = options[pos2]
optw3 = options[pos3]
print 'Optimal weight and corresponding realized gains: '
print 'The first cointegrated vector: ', optw1, max(output1)
print 'The second cointegrated vector: ', optw2, max(output2)
print 'The third cointegrated vector: ', optw3, max(output3)

# plot the relationship between realized gains and mixture weight for three cointegrated vectors
plt.plot(options, output1)
plt.plot(options, output2)
plt.plot(options, output3)
plt.legend(['first', 'second', 'third'], loc='best')
plt.xlabel('mixture weight')
plt.ylabel('Realized gains for three cointegrated vector on sampleB')
plt.title('The relationship between realized gains and mixture weight')
plt.show()
```

### Test Sample Processing

*Test the optimized signal in the Testing Sample C. Apply any of the signal quality metrics we discussed in class.*

```python
# extract vector element from sampleC
Y2 = sampleC.iloc[:,7]
Y3 = sampleC.iloc[:,11]
Y4= sampleC.iloc[:,15]
Y475 = sampleC.iloc[:,18]

# calculate signals1&2 in sampleC
CV1 = Y3-Y2*c1.beta[0]-c1.beta[1]
CV2 = Y4-Y3*c2.beta[0]-c2.beta[1]
CV3 = Y475-Y4*c3.beta[0]-c3.beta[1]
S1 = CV1 - CV1.ewm(span=55).mean()
S2 = CV2 - CV2.ewm(span=32).mean()
S3 = CV3 - CV3.ewm(span=31).mean()

# combine signals using optimal weight
d1 = optw1*CV1+(1-optw1)*S1
d2 = optw2*CV2+(1-optw2)*S2
d3 = optw3*CV3+(1-optw3)*S3
d_mer = [d1,d2,d3]

for j in range(3):
    d = d_mer[j]
    trading = [0,0] # trading signals
    forecast = [0,0]
    for i in range(2, len(d)-5):
        c = d[i]
        ar = ols(y=d[:i], x=d[:i].shift(1)) # build dynamic AR(1) model
        for _ in range(5): # predict rates in five days
            c = c*ar.beta[0]+ar.beta[1]
        if c-d[i]>0.00001: # If |Expected Gains(t)|>0.1 basis point, your trade is initiated
            trading.append(1)
            forecast.append(c-d[i])
        else:
            trading.append(0)
            forecast.append(0)

    realize = [] # store daily gain in a list
    for i in range(len(trading)):
        realize.append(trading[i]*(d[i+5]-d[i]))
    realize_cumsum = np.cumsum(realize) # calculate the cumsum of daily gains
    
    # corr between forecast and realized return
    corr = np.corrcoef(forecast, realize)[0][1]
    # % of time predicted correctly
    win_pct = float(len([item for item in realize if item > 0])) / len([item for item in trading if item ==1])
    print 'corr between realized and forecasted gains for d%s: '%str(j+1), corr
    print 'percentage of time predicted correctly for d%s: '%str(j+1), win_pct

    plt.plot(realize_cumsum) # plot the gains of specific example
    plt.title('Optimal realized gains on test sample for d%s'%str(j+1))
    plt.xlabel('days')
    plt.ylabel('Gains')
    plt.show()
```

### Conclusion

To conclude, the result in test sample is not as good as that in cross-validation sample. 

I think one of the most important reason is that the weight of cointegrated vector used in both cross-validation sample and test sample comes from training sample, which is probably outdated and not suitbale for current market. In this case, the so-called cointegrated vectors built in test sample are not actually cointegrated at all, and therefore the AR(1) models built on them are not robust, which may undermine the corresponding trading profit.

What's more, AR(1) may not be adequate to model the cointegrated vector since it has conditional variances, which calls for ARCH/GARCH.

From the result above for three cointegrated vectors, the first cointegrated vector got the best result and the performance of the second/third two vectors are not good. This means 2Y & 3Y have better cointegrated relationships and have more trading opportunities.

The winning percentage could be raised if we increase the trading threshold (currently 0.1 basic point).

The correlation between forecast and realized gains is far from satisfactory, which may be increased by improvement of model (AR(1) to ARIMA/GARCH) and frequate parameter updates.
