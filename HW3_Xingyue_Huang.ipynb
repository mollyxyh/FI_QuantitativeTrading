```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import quandl
import datetime

#quandl.ApiConfig.api_key = "QUANDLKEY"
#token = "QUANDLKEY"
#n = range(1,21)
#nms = ["CHRIS/CME_ED"+str(i) for i in n]
#dfs = [quandl.get(nm, authtoken=token) for nm in nms]

# perform data preprocessing
data = pd.read_csv('closes_Quandl_Rolling_EDs.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data = data.iloc[1251:, :] # select data from 2005 on

imm = pd.read_csv('imm_Quandl_Rolling_EDs.csv')
imm = imm.iloc[1251:len(imm)/2,:21] # select the first half part of imm data from 2005 on
for col in imm.columns:
    imm[col] = pd.to_datetime(imm[col]) # convert str dates to datetime format
imm.set_index('0', inplace=True) # congruent index
imm.columns = data.columns # congruent column names

assert imm.shape == data.shape, 'Warning: Two csv files have different dimensions!'
```

```python
# calculate actual/constant time to maturity and interpolated yields
actual = pd.DataFrame(columns = imm.columns, index = imm.index) # store actual time to maturity
constant = pd.DataFrame(columns = imm.columns, index = imm.index) # store constant time to maturity
yields = pd.DataFrame(columns = imm.columns, index = imm.index) # store linearly interpolated yields

for col in imm.columns:
    actual[col] = imm[col] - imm.index
actual = actual.applymap(lambda x: x.days)

for i in range(constant.shape[0]):
    for j in range(constant.shape[1]):
        y = constant.index[i].year
        m = constant.index[i].month
        d = constant.index[i].day
        y_add = (m+3*(j+1))/12
        m_new = (m+3*(j+1))%12
        if m_new == 0:
            y_add -=1
            m_new = 12
        try: 
            constant.iloc[i, j] = datetime.datetime(y+y_add, m_new, d) - constant.index[i]
        except ValueError:
            try:
                constant.iloc[i, j] = datetime.datetime(y+y_add, m_new, d-1) - constant.index[i]
            except ValueError:
                try:
                    constant.iloc[i, j] = datetime.datetime(y+y_add, m_new, d-2) - constant.index[i]
                except ValueError:
                    try:
                        constant.iloc[i, j] = datetime.datetime(y+y_add, m_new, d-3) - constant.index[i]
                    except ValueError:
                        constant.iloc[i, j] = datetime.datetime(y+y_add, m_new, d-4) - constant.index[i]
constant = constant.applymap(lambda x: x.days)

yields = pd.DataFrame(columns = imm.columns, index = imm.index)
for i in range(yields.shape[0]):
    for j in range(yields.shape[1]-1): # perform linear interpolation
        yields.iloc[i,j] = np.interp(constant.iloc[i,j], actual.iloc[i,:], data.iloc[i,:])
```

```python
# display actual/constant time to maturity & interpolated yields
print 'actual time to maturity: \n', actual.head(), '\n'
print 'constant time to maturity: \n', constant.head(), '\n'
print 'interpolated yields: \n', yields.head(), '\n'
yields.to_csv('DataSet3.csv') # save DataSet3.csv
```

```python
# plot yield curves for 2017-03-13 & 2017-03-14
x = np.linspace(0.25, 4.75, 20)
plt.plot(x, yields.iloc[-17,:].values, marker='o', label='interpolated yields for 2017-03-13')
plt.plot(x, yields.iloc[-16,:].values, marker='o', label='interpolated yields for 2017-03-14')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Interpolated yields on 2017-03-13 vs on 2017-03-14')
plt.xlabel('Years to maturity')
plt.ylabel('Interpolated yields')
plt.show()
```

```python
# calculate daily changes for dataset1 & dataset3
assert data.shape == yields.shape, 'Warnings: Two datasets of yields have different dimensions!'
roll_change = data.diff()
constant_change = yields.diff()
print 'Daily change (including roll): \n', roll_change.head(), '\n'
print 'Daily change in constant maturity futures: \n', constant_change.head()
```

```python
# plot constant & roll daily changes on 2017-03-13
plt.plot(x, roll_change.iloc[-17,:].values, marker='o', label='roll change')
plt.plot(x, constant_change.iloc[-17,:].values, marker='o', label='constant change')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Roll change vs constant change on 2017-03-13')
plt.xlabel('Time')
plt.ylabel('Years to maturity')
plt.show()
```

```python
# calculate & display volatility of two datasets
std_data1 = []
std_data3 = []
for col in data.columns:
    std_data1.append(data[col].std())
    std_data3.append(yields[col].std())
print 'std of dataset1: \n', std_data1, '\n'
print 'std of dataset3: \n', std_data3

plt.plot(std_data1, label='dataset1')
plt.plot(std_data3, label='dataset3')
plt.title('std of different datasets')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('tickers')
plt.ylabel('std')
plt.show()
```
